1. Program to Print Prime Numbers from a List (Without Built-in Functions)

```python
numbers = [10, 3, 5, 7, 12, 13, 17, 20]

for num in numbers:
    if num > 1:
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, "is prime")
```

3. User-defined Function (Count Even Numbers & Sum of Odd Numbers)

```python
def analyze_numbers(num_list):
    even_count = 0
    odd_sum = 0

    for num in num_list:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_sum += num

    return even_count, odd_sum


numbers = [1, 2, 3, 4, 5, 6]
even_count, odd_sum = analyze_numbers(numbers)

print("Count of even numbers:", even_count)
print("Sum of odd numbers:", odd_sum)
```

4. Function to Calculate Simple Interest (Default Arguments)

```python
def simple_interest(principal, rate=5, time=1):
    return (principal * rate * time) / 100


# Positional arguments
si1 = simple_interest(1000, 10, 2)
print("Simple Interest (Positional):", si1)

# Keyword arguments
si2 = simple_interest(principal=2000, rate=8, time=3)
print("Simple Interest (Keyword):", si2)

# Using default values
si3 = simple_interest(1500)
print("Simple Interest (Default):", si3)
```

5. Student Class with Grade Calculation

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "Fail"


# Creating objects
student1 = Student("Rahul", 85)
student2 = Student("Anita", 45)

print(student1.name, "Grade:", student1.calculate_grade())
print(student2.name, "Grade:", student2.calculate_grade())
```

6. BankAccount Class with Private Balance

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance


# Creating object
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)
print("Current Balance:", account.get_balance())
```

7. Division Program with Exception Handling

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

except ValueError:
    print("Error: Invalid input. Please enter numbers only.")
```

8. Custom Exception for Invalid Age

```python
class AgeError(Exception):
    pass

try:
    age = int(input("Enter age: "))

    if age < 0 or age > 120:
        raise AgeError("Age must be between 0 and 120.")

    print("Valid age:", age)

except AgeError as e:
    print("Error:", e)

except ValueError:
    print("Error: Please enter a valid number.")
```

9. File Handling (Create, Write, Read Student Data)

```python
# Create and write to file
file = open("student.txt", "w")
name = input("Enter student name: ")
marks = input("Enter marks: ")

file.write("Name: " + name + "\n")
file.write("Marks: " + marks + "\n")
file.close()

# Read and display file content
file = open("student.txt", "r")
content = file.read()
print("\nFile Content:")
print(content)
file.close()
```

10. Read Numbers from File and Calculate Total & Average

```python
try:
    file = open("numbers.txt", "r")
    numbers = file.readlines()
    file.close()

    total = 0
    count = 0

    for num in numbers:
        total += float(num.strip())
        count += 1

    average = total / count

    print("Total:", total)
    print("Average:", average)

except FileNotFoundError:
    print("Error: File does not exist.")
```
